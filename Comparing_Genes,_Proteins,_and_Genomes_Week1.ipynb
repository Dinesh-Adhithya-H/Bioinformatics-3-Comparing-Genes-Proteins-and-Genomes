#Changing money using dynamic programming

```python
money=18701
coins=[23,14,5,3,1]
```

```python
minnumcoins=[1000000 for i in range(money+1)] # technically set to infinity
minnumcoins[0]=0
for m in range(1,money+1):
  for i in range(len(coins)):
    if m>=coins[i]:
      if minnumcoins[m-coins[i]]<minnumcoins[m]:
        minnumcoins[m]=minnumcoins[m-coins[i]]+1
print(minnumcoins[money])
```

```python
i=money
shortest_path=[]
while i!=0:
  min=100000000000000
  last_j=0
  for j in coins:
    if minnumcoins[i-j]<min:
      min=minnumcoins[i-j]
      last_j=j
  i=i-last_j
  shortest_path.append(last_j)
```

#Longest Common Subsequence Problem

```python
def alpha(x,y):
  if x==y:
    return 1
  else:
    return 0
```

```python
def Needleman_Wunsch(seq1,seq2,n,m):
  table=[ [0 for i in range(m+1)] for j in range(n+1)]
  for i in range(1,n+1):
    for j in range(1,m+1):
      if alpha(seq1[i-1],seq2[j-1])==1:
        table[i][j]=table[i-1][j-1]+1
      else:
        table[i][j]=max(table[i-1][j],table[i][j-1])
  print("--")
  s1=''
  s2=''
  i=n
  j=m
  while i>0 and j>0:
    print(i,j)
    if table[i][j]==table[i-1][j]:
      s1="-"+s1
      s2=seq2[j-1]+s2
      i=i-1
    elif table[i][j]==table[i][j-1]:
      s2="-"+s2
      s1=seq1[i-1]+s1
      j=j-1
    else :
      s2=seq2[j-1]+s2
      s1=seq1[i-1]+s1
      i=i-1
      j=j-1
  return(s1,s2)
```

```python
nodes=[end]
while len(nodes)!=0:
  new_node=[]
  for i in nodes:
    for j in graph:
      if j[1]==i:
        new_node.append(j[0])
  print(nodes)
  nodes=new_node
```

```python
lengths=[0 for i in range(len(graph))]
nodes=[start]
while len(nodes)!=0:
  new_node=[]
  for j in nodes:
    for i in graph:
      if i[0]==j:
        if lengths[i[1]]<lengths[i[0]]+i[2]:
          lengths[i[1]]=lengths[i[0]]+i[2]
        new_node.append(i[1])
  nodes=new_node
```
